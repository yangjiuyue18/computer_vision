```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 读取图片
image = cv.imread('tlights/100_24.jpg')

# 转换为HSV颜色空间
img_hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)

# 黄点的中心大致位置 (x, y)
x, y = 50, 50  # 根据图片调整这个坐标

# 获取该像素的 HSV 值
h, s, v = img_hsv[y, x]

# 在原图上标记出该位置的黑点
marked_image = image.copy()
cv.circle(marked_image, (x, y), radius=10, color=(0, 0, 0), thickness=-1)  # 在 (x, y) 位置绘制黑点

# 转换为 RGB 格式以便用 matplotlib 显示
marked_image_rgb = cv.cvtColor(marked_image, cv.COLOR_BGR2RGB)

# 使用 matplotlib 显示图片
plt.imshow(marked_image_rgb)
plt.title(f"HSV at ({x}, {y}) - H: {h}, S: {s}, V: {v}")
plt.axis('off')  # 不显示坐标轴
plt.show()

# 同时展示 HSV 值
h, s, v
```
